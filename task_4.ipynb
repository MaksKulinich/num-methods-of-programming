```python
a = np.array([[1,2,3],[1,3,4],[2,-1,-1]])
b = np.array([[5],[6],[1]])
def kram(a,b):# метод Крамера
    det_a = np.linalg.det(a)
    if det_a != 0:
       a1 = np.matrix(a)
       a2 = np.matrix(a)
       a3 = np.matrix(a)
       a1[:, 0] = b
       a2[:, 1] = b
       a3[:, 2] = b
       x = np.linalg.det(a1)/det_a
       y =np.linalg.det(a2)/det_a
       z= np.linalg.det(a3)/det_a
       # print('Solving a system of linear algebraic equations by Cramer*s method')
       print('x=',x.round(1),'y=',y.round(1),'z=',z.round(1)) 
    else:
        print("Система немає розв'язків, оскільки визначник дорівнює 0 ")
    return(x,y,z)
kram(a,b)
print("Перевірка розв'язання : ","\n", np.linalg.solve(a,b).round(1))
```

```python
import random as rd
n = 3
m = 4
a = np.zeros((n, m), dtype=int)
for i in range(n): #наповнюємо матрицю випадковими числами від 1 до 20
    for j in range(m):
        a[i][j] = rd.randint(1, 20)

print(a)

for i in range(n): #обчислення середнього значення для кожного рядка
    suma_row = 0
    for j in range(m):
        suma_row += a[i][j]
    average_row = suma_row / m
    print(f"Середнє значення {i+1} рядка: {(average_row).round(1)}")

for j in range(m): #обчислення середнього значення для кожного стовпця
    suma_column = 0
    for i in range(n):
        suma_column += a[i][j]
    average_column = suma_column / n
    print(f"Середнє значення {j+1} стовпця: {(average_column).round(1)}")
```
